```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
features = pd.read_csv('/content/city_temperature.csv')
features.head(5)
```

```python
features=features.drop("State",axis=1)
```

```python
features=features.drop([1111021,1152561])
```

```python
features.info()
```

```python
features
```

```python
X = features.iloc[:, 3:6].values
y = features.iloc[:, 6].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators = 10, random_state = 42)
rf.fit(X_train,  y_train);
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (18,8))
plt.plot(y_test[0:200],label ="Predicted",linewidth='0.5')
plt.plot(y_test[0:200],label = "Actual",linewidth='0.5')
plt.legend()
plt.show()
```
